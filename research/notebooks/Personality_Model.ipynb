````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Separate features (X) and target (y)
X = df_filtered.drop('Unified_Role', axis=1)
y = df_filtered['Unified_Role']

# Encode the target variable 'Role'
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f"Original class distribution: {Counter(y_encoded)}")

# Visualize the original distribution
plt.figure(figsize=(12, 7))
sns.countplot(y=y_encoded, order = pd.Series(y_encoded).value_counts().index, palette='viridis')
plt.title('Distribution of Encoded Job Roles (Original)')
plt.xlabel('Count')
plt.ylabel('Encoded Role')
plt.show()

print("Note: The y-axis shows encoded role labels. You can map them back to original role names using 'le.inverse_transform'.")
````

````python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Initialize the StandardScaler
scaler = StandardScaler()

# Scaling will be applied after the train-test split to prevent data leakage.
# The scaler will be fitted only on the training data.
print("StandardScaler initialized. It will be fitted and applied after the train-test split.")
````

````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE # Import SMOTE here
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler # Ensure StandardScaler is imported if not globally available

# Assuming X and y_encoded are in scope from cell e60e9b2d

# 1. Split the data into raw training, validation, and testing sets
x_train_raw, X_temp_raw, y_train_raw, y_temp_raw = train_test_split(X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)
x_val_raw, x_test_raw, y_val_raw, y_test_raw = train_test_split(X_temp_raw, y_temp_raw, test_size=0.5, random_state=42, stratify=y_temp_raw)

# 2. Apply standardization
# Fit scaler only on the training data and transform all sets
x_train = scaler.fit_transform(x_train_raw)
x_val = scaler.transform(x_val_raw)
x_test = scaler.transform(x_test_raw)

# Convert to DataFrame to maintain column names for consistency
x_train = pd.DataFrame(x_train, columns=X.columns)
x_val = pd.DataFrame(x_val, columns=X.columns)
x_test = pd.DataFrame(x_test, columns=X.columns)

# # 3. Apply SMOTE to the *training* data only
smote = SMOTE(random_state=42)
x_train_smoted, y_train_smoted = smote.fit_resample(x_train, y_train_raw) # Use y_train_raw as target for smote
# x_train_smoted, y_train_smoted = x_train, y_train_raw # Use y_train_raw as target for smote

# Convert y_train_smoted, y_val_raw, y_test_raw to Series for consistency
y_train_smoted = pd.Series(y_train_smoted, name='Role_encoded')
y_val = pd.Series(y_val_raw, name='Role_encoded') # Ensure y_val is a Series
y_test = pd.Series(y_test_raw, name='Role_encoded') # Ensure y_test is a Series

print(f"Original Training features shape: {x_train_raw.shape}")
print(f"SMOTEd Training features shape: {x_train_smoted.shape}")
print(f"Validation features shape: {x_val.shape}")
print(f"Testing features shape: {x_test.shape}")
print(f"Original Training target shape: {y_train_raw.shape}")
print(f"SMOTEd Training target shape: {y_train_smoted.shape}")
print(f"Validation target shape: {y_val.shape}")
print(f"Testing target shape: {y_test.shape}")

# Visualize the distribution after SMOTE (training data only)
plt.figure(figsize=(12, 7))
sns.countplot(y=y_train_smoted, order = y_train_smoted.value_counts().index, palette='viridis')
plt.title('Distribution of Encoded Job Roles After SMOTE (Training Data)')
plt.xlabel('Count')
plt.ylabel('Encoded Role')
plt.show()

print("Note: The y-axis shows encoded role labels. You can map them back to original role names using 'le.inverse_transform'.")
````

````python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score

baselines = {
    'RandomForest':       RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1),
    'GradientBoosting':   GradientBoostingClassifier(n_estimators=100, random_state=42),
    'XGBoost':            XGBClassifier(n_estimators=200, use_label_encoder=False,
                                        eval_metric='mlogloss', random_state=42, n_jobs=-1),
    'LightGBM':           LGBMClassifier(n_estimators=200, random_state=42, n_jobs=-1, verbose=-1),
}

results = {}
for name, clf in baselines.items():
    clf.fit(x_train_smoted, y_train_smoted)
    val_acc  = accuracy_score(y_val, clf.predict(x_val))
    test_acc = accuracy_score(y_test, clf.predict(x_test))
    val_f1   = f1_score(y_val, clf.predict(x_val), average='weighted')
    results[name] = {'Val Acc': val_acc, 'Test Acc': test_acc, 'Val F1': val_f1}
    print(f"{name:25s} | Val Acc: {val_acc:.4f} | Test Acc: {test_acc:.4f} | Val F1: {val_f1:.4f}")

results_df = pd.DataFrame(results).T
results_df
````

````python
import optuna
from optuna.samplers import TPESampler
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Silence optuna logs if you want a clean terminal
optuna.logging.set_verbosity(optuna.logging.WARNING)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

def lgbm_objective(trial):
    params = {
        'n_estimators':       trial.suggest_int('n_estimators', 100, 500, step=50),
        'max_depth':          trial.suggest_int('max_depth', 3, 12),
        'learning_rate':      trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'num_leaves':         trial.suggest_int('num_leaves', 20, 150),
        'subsample':          trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree':   trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'min_child_samples':  trial.suggest_int('min_child_samples', 5, 50),
        'reg_alpha':          trial.suggest_float('reg_alpha', 0.0, 1.0),
        'reg_lambda':         trial.suggest_float('reg_lambda', 0.0, 2.0),
        'random_state':       42,
        'n_jobs':             -1,
        'verbose':            -1,
    }
    clf = LGBMClassifier(**params)
    score = cross_val_score(
        clf, x_train_smoted, y_train_smoted,
        cv=cv, scoring='f1_weighted', n_jobs=-1
    ).mean()
    return score

print("Running Optuna tuning for LightGBM (30 trials, max 5 min)...")
lgbm_study = optuna.create_study(
    direction='maximize',
    sampler=TPESampler(seed=42)
)
lgbm_study.optimize(lgbm_objective, n_trials=30, timeout=300)

print(f"\nBest CV F1 (weighted): {lgbm_study.best_value:.4f}")
print(f"Best params:           {lgbm_study.best_params}")
````

````python
best_lgbm = LGBMClassifier(
    **lgbm_study.best_params,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1,
    verbose=-1
)
best_lgbm.fit(x_train_smoted, y_train_smoted)
````

````python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.frozen import FrozenEstimator

calibrated_lgbm = CalibratedClassifierCV(
    FrozenEstimator(best_lgbm),
    method='isotonic'
)
calibrated_lgbm.fit(x_val, y_val)
````

````python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred  = calibrated_lgbm.predict(x_test)
y_proba = calibrated_lgbm.predict_proba(x_test)

print("\n" + "="*60)
print("TUNED LIGHTGBM — FINAL EVALUATION")
print("="*60)
print(f"Test Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"Test F1 (wtd) : {f1_score(y_test, y_pred, average='weighted'):.4f}")
print()
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(12, 9))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix — Tuned LightGBM (Calibrated)')
plt.xlabel('Predicted'); plt.ylabel('Actual')
plt.xticks(rotation=30, ha='right'); plt.tight_layout(); plt.show()

# Confidence distribution
max_probs = y_proba.max(axis=1)
plt.figure(figsize=(9, 4))
plt.hist(max_probs, bins=25, color='steelblue', edgecolor='white')
plt.axvline(max_probs.mean(), color='red', linestyle='--',
            label=f'Mean: {max_probs.mean():.2f}')
plt.title('Prediction Confidence Distribution (Calibrated LightGBM)')
plt.xlabel('Max Predicted Probability'); plt.ylabel('Count')
plt.legend(); plt.tight_layout(); plt.show()

print(f"\nMean confidence   : {max_probs.mean():.3f}")
print(f"% above 0.70 conf : {(max_probs > 0.70).mean()*100:.1f}%")

# Compare baseline vs tuned
print("\n" + "="*60)
print("BASELINE vs TUNED COMPARISON")
print("="*60)
baseline_lgbm = baselines['LightGBM']
print(f"Baseline LightGBM — Test Acc: 0.8042 | Test F1: {f1_score(y_test, baseline_lgbm.predict(x_test), average='weighted'):.4f}")
print(f"Tuned    LightGBM — Test Acc: {accuracy_score(y_test, y_pred):.4f} | Test F1: {f1_score(y_test, y_pred, average='weighted'):.4f}")
````

# Career Mapping Project Summary

This notebook details the process of building and deploying a machine learning model to predict career roles based on skills and personality traits.

## 1. Data Loading and Initial Exploration

The project begins by loading the `CareerMapping1.csv` dataset. Initial inspection revealed 15 columns, including 14 numerical features (skills and personality traits) and one categorical target feature ('Role'). The dataset contained approximately 10,179 entries before duplicate removal. No missing values were found.

- **Dataset Source:** `CareerMapping1.csv`
- **Initial Columns:** 15 (14 features, 1 target)
- **Initial Records:** 10,179
- **Missing Values:** None

Duplicate entries were identified and removed, reducing the dataset size to 4,116 unique records. A correlation matrix was generated to understand the relationships between numerical features.

## 2. Data Preprocessing

### Unified Role Mapping
To simplify the classification task and potentially address class imbalance, the original 16 detailed job roles were mapped into 6 broader 'Unified_Role' categories. This step helps in creating more robust and generalized predictions.

- **Original Roles:** 16 (e.g., 'Software Developer', 'AI ML Specialist')
- **Unified Roles:** 6 (e.g., 'Artificial Intelligence', 'Security', 'Software Development and Engineering')

### Label Encoding
The 'Unified_Role' target variable was then label encoded into numerical representations for model training.

### Data Splitting and Scaling
The dataset was split into training, validation, and testing sets using a 70/15/15 ratio, ensuring stratification to maintain the class distribution in each split. `StandardScaler` was used to standardize the numerical features, preventing data leakage by fitting the scaler only on the training data and then transforming all splits.

- **Split Ratio:** Training (70%), Validation (15%), Testing (15%)
- **Scaling:** `StandardScaler` applied to features.

### Handling Class Imbalance with SMOTE
An initial analysis of the unified role distribution showed a significant class imbalance. To mitigate this, Synthetic Minority Over-sampling Technique (SMOTE) was applied *only* to the training data. This technique generates synthetic samples for minority classes, balancing the dataset and improving the model's ability to learn from all classes.

- **Method:** SMOTE (Synthetic Minority Over-sampling Technique)
- **Application:** Applied exclusively to the training data (`x_train_raw`, `y_train_raw`).
- **Result:** The training data was re-sampled to have a balanced distribution across all unified roles.

## 3. Model Training and Evaluation (Baselines)

Several baseline classification models were trained and evaluated on the SMOTEd training data and assessed on the raw validation and test sets:

- `RandomForestClassifier`
- `GradientBoostingClassifier`
- `XGBoostClassifier`
- `LGBMClassifier`

The `LightGBM` model showed promising performance, particularly in terms of Test Accuracy and F1-weighted score, making it a strong candidate for further optimization.

| Model             | Val Acc | Test Acc | Val F1 |
|-------------------|---------|----------|--------|
| RandomForest      | 0.7942  | 0.7994   | 0.7947 |

## 4. Hyperparameter Tuning (LightGBM)

Optuna, an automatic hyperparameter optimization framework, was used to fine-tune the `LGBMClassifier`. The objective was to maximize the weighted F1-score through 30 trials over a maximum of 5 minutes. The best parameters identified by Optuna were then used to initialize the `best_lgbm` model.

- **Optimization Tool:** Optuna
- **Objective:** Maximize `f1_weighted` score via 5-fold StratifiedKFold cross-validation.
- **Trials:** 30

## 5. Model Calibration

To improve the reliability of the predicted probabilities, the `best_lgbm` model was calibrated using `CalibratedClassifierCV` with `method='isotonic'` on the validation set. This step ensures that the predicted probabilities align more closely with the true likelihoods, which is crucial for applications where confidence in predictions is important.

## 6. Final Model Evaluation

The calibrated LightGBM model was rigorously evaluated on the test set. Key metrics included:

- **Test Accuracy:** 0.8074
- **Test F1 (weighted):** 0.8098

The classification report provided detailed precision, recall, and F1-scores for each unified role, highlighting the model's performance across different classes. A confusion matrix visually summarized the correct and incorrect classifications.

Additionally, the distribution of prediction confidences was analyzed, showing that the mean confidence was approximately 0.823, with 74.9% of predictions having a confidence above 0.70. This indicates a good level of certainty in the model's predictions after calibration.

The tuned and calibrated LightGBM model demonstrated a slight improvement over the baseline LightGBM, affirming the benefits of hyperparameter tuning and calibration.
