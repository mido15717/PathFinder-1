```python
# ─────────────────────────────────────────────────────────────────────────────
# SKILL_GAP_MAP  — used ONLY for skill-gap output, NOT for model training.
#
# Structure:
#   career_name -> {
#       feature_group -> {
#           'ordinal_label': human-readable name,
#           'levels': [  (score_value, [skills that define this level])  ]
#       }
#   }
#
# For the mobile example:
#   android only  → score 1
#   ios only      → score 2
#   both/general  → score 3
#   missing       → score 0  → gap = needs mobile skills
# ─────────────────────────────────────────────────────────────────────────────

SKILL_GAP_MAP = {
    "mobile_development": {
        "ordinal_label": "Mobile Development Coverage",
        "levels": [
            (0, [],                                        "No mobile skills"),
            (1, ["Android App Development"],               "Android only"),
            (2, ["iOS App Development"],                   "iOS only"),
            (3, ["Mobile App Development"],                "Cross-platform / general"),
        ]
    },
    "web_stack": {
        "ordinal_label": "Web Stack Coverage",
        "levels": [
            (0, [],                                        "No web skills"),
            (1, ["Web Development"],                       "General web"),
            (2, ["Web Development", "HTML/CSS"],           "HTML/CSS added"),
            (3, ["Web Development", "HTML/CSS",
                 "JavaScript"],                           "Full frontend stack"),
        ]
    },
    "security_depth": {
        "ordinal_label": "Security Depth",
        "levels": [
            (0, [],                                        "No security skills"),
            (1, ["Cyber Security"],                        "General cybersecurity"),
            (2, ["Cyber Security", "Network Security"],    "+ Network security"),
            (3, ["Cyber Security", "Network Security",
                 "Cryptography"],                         "Full security stack"),
        ]
    },
    "ai_depth": {
        "ordinal_label": "AI/ML Depth",
        "levels": [
            (0, [],                                        "No AI skills"),
            (1, ["Artificial Intelligence and Machine Learning"],
                                                           "General AI/ML"),
            (2, ["Artificial Intelligence and Machine Learning",
                 "Deep Learning"],                        "+ Deep Learning"),
            (3, ["Artificial Intelligence and Machine Learning",
                 "Deep Learning", "Natural Language Processing"],
                                                           "Full AI stack"),
        ]
    },
    "data_science_depth": {
        "ordinal_label": "Data Science Depth",
        "levels": [
            (0, [],                                        "No data skills"),
            (1, ["Data Analysis and Visualization"],       "Data analysis"),
            (2, ["Data Analysis and Visualization",
                 "Statistical Analysis"],                 "+ Statistics"),
            (3, ["Data Analysis and Visualization",
                 "Statistical Analysis", "Data Engineering"],
                                                           "Full data stack"),
        ]
    },
}


def get_skill_gap(skill_string: str, predicted_career: str,
                  career_required_skills: dict) -> dict:
    """
    Returns a dict of what the student is missing for their predicted career.
    career_required_skills: {career -> set of skills expected for that career}
    Built from the training data (see Cell 52 below).
    """
    student = {s.strip() for s in skill_string.split(',') if s.strip()}
    expected = career_required_skills.get(predicted_career, set())
    missing = expected - student

    # Also compute ordinal scores for each skill group
    ordinal_scores = {}
    for group, meta in SKILL_GAP_MAP.items():
        score = 0
        for level_score, level_skills, level_label in meta['levels']:
            if level_skills and all(s in student for s in level_skills):
                score = level_score
        ordinal_scores[group] = score

    return {
        'missing_skills': sorted(missing),
        'ordinal_scores': ordinal_scores
    }

print('SKILL_GAP_MAP defined.')
```

```python
def score_row(skill_string: str) -> dict:
    """
    Converts a comma-separated skill string into a feature dict.
    Score per feature = count of sub-skills the student has in that group.
    f_sparse_row = 1 if student has fewer than 2 skills (low-quality input).
    """
    if pd.isna(skill_string) or not str(skill_string).strip():
        row = {f: 0 for f in FEATURE_NAMES}
        row['f_sparse_row'] = 1
        return row

    student_skills = {s.strip() for s in str(skill_string).split(',') if s.strip()}

    scores = {}
    for feature, sub_skills in FEATURE_SCHEMA.items():
        if feature == 'f_sparse_row':
            continue
        scores[feature] = sum(1 for s in sub_skills if s in student_skills)

    # Sparse row flag — set AFTER counting so it doesn't affect skill counts
    scores['f_sparse_row'] = 1 if len(student_skills) < 2 else 0

    return scores


# Quick sanity check
test = "Android App Development, Deep Learning, Cyber Security"
result = score_row(test)
non_zero = {k: v for k, v in result.items() if v > 0}
print(f"Test input: {test}")
print(f"Non-zero features: {non_zero}")
print(f"Sparse flag: {result['f_sparse_row']}  (expected 0)")

sparse_test = score_row("Web Development")
print(f"\nSingle-skill sparse flag: {sparse_test['f_sparse_row']}  (expected 1)")
```

```python
print("Loading dataset...")
df2 = pd.read_excel("Career Dataset - Standardized.xlsx")
print(f"  Raw rows: {len(df2)}  |  Careers: {df2['Career'].nunique()}")
print(f"  Career distribution:\n{df2['Career'].value_counts()}\n")

# ── Remove sparse rows (fewer than 2 skills) ──────────────────────────────
# These rows have almost no signal and hurt model quality.
# 703 rows affected — ~17% of data, but very low information rows.
df2['_skill_count'] = df2['Skill'].apply(
    lambda x: len([s for s in str(x).split(',') if s.strip()])
)
df2_clean = df2[df2['_skill_count'] >= 2].reset_index(drop=True)
print(f"  After removing sparse rows (<2 skills): {len(df2_clean)} rows")
print(f"  Removed: {len(df2) - len(df2_clean)} rows")
print(f"  Career distribution after filter:\n{df2_clean['Career'].value_counts()}\n")

# ── Build feature matrix ──────────────────────────────────────────────────
print("Building feature matrix...")
X = pd.DataFrame(
    df2_clean['Skill'].apply(score_row).tolist(),
    columns=FEATURE_NAMES
)
print(f"  Feature matrix shape: {X.shape}")

# ── Verify: check features are actually doing something ───────────────────
print("\nFeature coverage (% of rows with score > 0):")
coverage = (X > 0).mean().sort_values(ascending=False)
for feat, pct in coverage.items():
    bar = '█' * int(pct * 30)
    print(f"  {feat:35s} {pct:5.1%}  {bar}")

# ── Build career required-skills profile (for skill gap analysis) ─────────
# For each career, find the skills present in >=40% of that career's students.
# These become the "expected" skills for skill-gap reporting.
print("\nBuilding career skill profiles for gap analysis...")
CAREER_SKILL_PROFILES = {}
for career in df2_clean['Career'].unique():
    sub = df2_clean[df2_clean['Career'] == career]
    all_s = []
    for s in sub['Skill']:
        all_s.extend([x.strip() for x in str(s).split(',') if x.strip()])
    from collections import Counter
    counts = Counter(all_s)
    n = len(sub)
    # Skills present in >=40% of students in this career
    expected = {skill for skill, cnt in counts.items() if cnt / n >= 0.40}
    CAREER_SKILL_PROFILES[career] = expected
    print(f"  {career}: {len(expected)} expected skills")

print("\nDone.")
```

## Train and Evaluate LogisticRegression

This section defines, trains, and evaluates a `LogisticRegression` classifier, including hyperparameter optimization using Optuna. Its performance is then evaluated and compared to the previous models.

```python
print('Re-executing Logistic Regression model setup with typo fixed.')
```

```python
import optuna
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.calibration import CalibratedClassifierCV

optuna.logging.set_verbosity(optuna.logging.WARNING);

# Reuse the same StratifiedKFold cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42);

def lr_objective(trial):
    # Suggest solver first
    solver = trial.suggest_categorical('solver', ['liblinear', 'lbfgs', 'saga'])

    # Suggest penalty from the *union* of all possible penalties ('l1', 'l2')
    # This ensures 'penalty' is always suggested with the same choices across trials,
    # and we handle incompatibility with explicit pruning below.
    penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])

    params = {
        'C': trial.suggest_float('C', 0.01, 10.0, log=True),
        'solver': solver,
        'penalty': penalty,
        'random_state': 42,
        'n_jobs': -1,
        'max_iter': 500 # Increased max_iter for convergence
    }

    # Explicit pruning for incompatible penalty/solver combinations
    # LBFGS only supports L2 penalty.
    if params['solver'] == 'lbfgs' and params['penalty'] == 'l1':
        raise optuna.exceptions.TrialPruned(f"Incompatible combination: solver={params['solver']}, penalty={params['penalty']}")
    # Other solvers (liblinear, saga) support both l1 and l2, so no pruning needed for them.

    clf_trial = LogisticRegression(**params)
    score = cross_val_score(
        clf_trial, X_train, y_train, cv=cv, scoring='f1_weighted', n_jobs=-1
    ).mean()
    return score

print("Running Optuna hyperparameter search for Logistic Regression (30 trials, max 5 min)...")
lr_study = optuna.create_study(
    direction='maximize',
    sampler=optuna.samplers.TPESampler(seed=42)
)
lr_study.optimize(lr_objective, n_trials=30, timeout=300)

print(f"\nBest Logistic Regression CV F1 (weighted): {lr_study.best_value:.4f}")
print(f"Best Logistic Regression params: {lr_study.best_params}")

# Train final Logistic Regression model with best params
best_lr_params = lr_study.best_params
best_lr_params.update({'random_state': 42, 'n_jobs': -1, 'max_iter': 500})

lr_clf = LogisticRegression(**best_lr_params)
lr_clf.fit(X_train, y_train)

# Calibrate the Logistic Regression classifier
lr_clf_calibrated = CalibratedClassifierCV(lr_clf, method='isotonic')
lr_clf_calibrated.fit(X_test, y_test)

print("\nLogistic Regression Model trained and calibrated.")
```

```python
def predict_career(skill_string: str,
                   confidence_threshold: float = 0.40) -> dict:
    """
    Given a comma-separated skill string from a user survey,
    returns:
      - predicted career
      - confidence score (calibrated probability)
      - top 3 careers with probabilities
      - skill gap report (what's missing vs predicted career)
      - ordinal scores per skill group
    """
    # Score the input
    features = score_row(skill_string)
    X_input = pd.DataFrame([features], columns=FEATURE_NAMES)

    # Predict
    probs = rf_clf_calibrated.predict_proba(X_input)[0]
    top3_idx = np.argsort(probs)[::-1][:3]
    top3 = [(le.classes_[i], round(float(probs[i]), 3)) for i in top3_idx]

    predicted_career = top3[0][0]
    confidence = top3[0][1]

    # Skill gap
    gap = get_skill_gap(skill_string, predicted_career, CAREER_SKILL_PROFILES)

    result = {
        'predicted_career': predicted_career if confidence >= confidence_threshold
                            else 'Uncertain — please provide more skills',
        'confidence': confidence,
        'top_3': top3,
        'missing_skills': gap['missing_skills'],
        'ordinal_scores': gap['ordinal_scores'],
        'input_quality': 'low' if features['f_sparse_row'] == 1 else 'ok'
    }
    return result


# ── Demo ─────────────────────────────────────────────────────────────────
demo_input = "Deep Learning, Natural Language Processing, Data Engineering, Statistical Analysis"
result = predict_career(demo_input)

print(f"Input:             {demo_input}")
print(f"Predicted career:  {result['predicted_career']}")
print(f"Confidence:        {result['confidence']:.1%}")
print(f"Top 3:             {result['top_3']}")
print(f"Missing skills:    {result['missing_skills']}")
print(f"Ordinal scores:    {result['ordinal_scores']}")
print(f"Input quality:     {result['input_quality']}")
```

```python
def run_precise_skills_survey():
    print("=" * 60)
    print("🎓 PRECISE CAREER RECOMMENDATION SURVEY (MODEL 2) 🎓")
    print("=" * 60)
    print("Select the EXACT skills you have to build an accurate profile.\n")

    selected_skills = []

    # ── STEP 1: Handle Tracks from SKILL_GAP_MAP ────────────────────────────
    print("--- 🗺️ CORE CAREER TRACK DEPTHS ---")
    for group_key, meta in SKILL_GAP_MAP.items():
        print(f"\n🔹 {meta['ordinal_label']}:")
        for level_score, level_skills, level_label in meta['levels']:
            # Clean up 'Web Development' or noise skills if they aren't in your schema
            clean_skills = [s for s in level_skills if s != "Web Development"]
            print(f"  [{level_score}] {level_label}")
            if clean_skills:
                print(f"      (Gives skills: {', '.join(clean_skills)})")

        max_level = max(level[0] for level in meta['levels'])
        while True:
            try:
                choice = int(input(f"👉 Enter choice (0-{max_level}): "))
                if 0 <= choice <= max_level:
                    for level_score, level_skills, _ in meta['levels']:
                        if level_score == choice:
                            # Filter out 'Web Development' noise string here too
                            selected_skills.extend([s for s in level_skills if s != "Web Development"])
                    break
                print(f"❌ Selection out of bounds.")
            except ValueError:
                print("❌ Enter a valid integer.")

    # ── STEP 2: Handle Specific Sub-Skills accurately ──────────────────────
    covered_skills = set(
        skill for meta in SKILL_GAP_MAP.values()
        for level in meta['levels'] for skill in level[1]
    )

    print("\n" + "─" * 40)
    print("--- 🛠️ ADDITIONAL TECH & SOFT SKILL GROUPS ---")
    print("Enter the numbers of ALL sub-skills you possess, separated by commas (e.g., 1,3,4).")
    print("If you have none of them, simply enter 0.")

    for feature_name, sub_skills in FEATURE_SCHEMA.items():
        if feature_name == 'f_sparse_row' or not sub_skills:
            continue

        display_skills = [s for s in sub_skills if s not in covered_skills]
        if not display_skills:
            continue

        clean_label = feature_name.replace("f_", "").replace("_", " ").title()
        print(f"\n🔸 {clean_label}:")
        for i, s in enumerate(display_skills, 1):
            print(f"  {i}. {s}")

        while True:
            user_input = input(f"👉 Select your skills (or 0 for none): ").strip()
            if user_input == '0':
                break
            try:
                # Parse comma-separated index choices
                choices = [int(x.strip()) for x in user_input.split(',') if x.strip()]
                if all(1 <= c <= len(display_skills) for c in choices):
                    for c in choices:
                        selected_skills.append(display_skills[c - 1])
                    break
                print(f"❌ Invalid choice. Please use numbers between 1 and {len(display_skills)}.")
            except ValueError:
                print("❌ Please enter numbers separated by commas (e.g., 1,2).")

    # ── STEP 3: Generate String and Clean Noise Artifacts ───────────────────
    unique_selected_skills = list(dict.fromkeys(selected_skills))
    final_skill_string = ", ".join(unique_selected_skills)

    print("\n" + "=" * 60)
    print("🔄 Processing your profile and running prediction metrics...")
    print("=" * 60)

    # Call your standard prediction function
    prediction_result = predict_career(final_skill_string)

    # Post-process 'missing_skills' to remove global noise strings like 'Adaptability'
    NOISE_SKILLS = {'Adaptability', 'Communication', 'Critical Thinking', 'Graphic Design', 'Web Development'}
    prediction_result['missing_skills'] = [
        s for s in prediction_result['missing_skills'] if s not in NOISE_SKILLS
    ]

    # ── STEP 4: Output Cleaned Results ──────────────────────────────────────
    print(f"\n🎯 FINAL PREDICTION")
    print(f"   Career         : {prediction_result['predicted_career']}")
    print(f"   Confidence     : {prediction_result['confidence']:.1%}")
    print(f"   Top 3          : {prediction_result['top_3']}")
    print(f"   Input quality  : {prediction_result['input_quality']}")
    print(f"   Missing skills : {prediction_result['missing_skills']}")
    print(f"   Current Tracks : {prediction_result['ordinal_scores']}")
    print("=" * 60 + "\n")

    return prediction_result

run_precise_skills_survey()
```
